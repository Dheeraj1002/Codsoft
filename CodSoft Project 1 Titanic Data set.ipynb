## Columns or Features Description are as follows Â¶
#### > PassengerId: A unique identifier assigned to each passenger.
#### > Survived: Indicates whether a passenger survived (1) or died (0).
#### > Pclass: Represents the passenger class, which can be a numerical or categorical value indicating the class of travel (e.g., first, second, or third class).

#### > Sex: Represents the gender of the passenger, typically categorized as male or female.
#### > Age: The age of the passenger, often represented as a numerical value.
#### > SibSp: Indicates the number of siblings or spouses traveling with the passenger.

#### > Ticket: The ticket number associated with the passenger's ticket.
#### > Fare: The amount of money spent on the ticket.
#### > Cabin: Represents the cabin category or number assigned to the passenger, if available.
#### > Embarked: Indicates the port where the passenger embarked from, typically represented by a single character code (e.g., C for Cherbourg, Q for Queenstown, S for Southampton).

## Handling missing Data Â¶

```python
# There is a lot of missing data on the Cabin column, We drop Cabin column .
# PassengerId, Name and Ticket column are not relevant to their survival chances hence, We will drop that too
# To fill in Age column we will use Mean of Age Column.
# To fill Embarked Column we will check with the data if they are related to other Passengers.
Titanic.isnull().sum()
```

```python
# Filling mean value for Age wherever data was missing
Titanic['Age'].fillna(Titanic['Age'].mean(), inplace=True)
```

```python
Titanic[Titanic['Embarked'].isnull()]
```

```python
# Checking the Ticket number that matches with other crew.
rows_of_113572_in_column = Titanic[Titanic['Ticket'].eq('113572')]
rows_of_113572_in_column
```

```python
# Checking the Cabin number that matches with other crew.
rows_of_B28_in_column = Titanic[Titanic['Cabin'].eq('B28')]
rows_of_B28_in_column
```

```python
# By the above data we get to the conclusion that these both passenger are related 
# There is no other relation so we will fill the Embarked column with 'U' as in unknown
Titanic['Embarked'].fillna('U' ,inplace = True) 
```

```python
# Dropping the following columns
Titanic = Titanic.drop(['PassengerId','Name','Cabin','Ticket'], axis = 1 )
```

```python
Titanic.isnull().sum()
```

## Encoding Â¶

```python
Titanic.head()
```

```python
Titanic['Sex'] = Titanic['Sex'].replace('male', 0)
Titanic['Sex'] = Titanic['Sex'].replace('female', 1)
Titanic['Embarked'] = Titanic['Embarked'].replace('S', 0)
Titanic['Embarked'] = Titanic['Embarked'].replace('C', 1)
Titanic['Embarked'] = Titanic['Embarked'].replace('Q', 2)
Titanic['Embarked'] = Titanic['Embarked'].replace('U', 3)
```

## Feature Scaling Â¶

```python
x = Titanic.drop(columns = ['Survived'])
y = Titanic['Survived']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
y = y.to_numpy()
```

```python
x
```

```python
y
```

```python
# Splitting the Data into Train and Test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state= 0)
```

## Building the Logistic Regression Model Â¶

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
## Prediction Accuracy and Classification report
from sklearn.metrics import accuracy_score, classification_report
pred_xtrain = model.predict(x_train)
accuracy_xtrain = accuracy_score(y_train,pred_xtrain)
print('accuracy_xtrain :', accuracy_xtrain)
```

```python
from sklearn.metrics import accuracy_score, classification_report
pred_xtest = model.predict(x_test)
accuracy_xtest = accuracy_score(y_test,pred_xtest)
print('accuracy_xtest :', accuracy_xtest)
```

```python
print(classification_report(y_test,pred_xtest))
```

## Building the RandomForestClassifier Model Â¶

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()
model1.fit(x_train, y_train)
```

```python
## Prediction Accuracy and Classification report
from sklearn.metrics import accuracy_score, classification_report
pred_xtrain = model1.predict(x_train)
accuracy_xtrain = accuracy_score(y_train,pred_xtrain)
print('accuracy_xtrain :', accuracy_xtrain)
```

```python
from sklearn.metrics import accuracy_score, classification_report
pred_xtest = model1.predict(x_test)
accuracy_xtest = accuracy_score(y_test,pred_xtest)
print('accuracy_xtest :', accuracy_xtest)
```

```python
print(classification_report(y_test,pred_xtest))
```
